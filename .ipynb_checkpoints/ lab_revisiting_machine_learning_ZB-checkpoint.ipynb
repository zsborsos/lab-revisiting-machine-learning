Identify columns that over 85% missing values

```python
null_percent_categorical = pd.DataFrame((categorical_data.isnull().sum()/len(categorical_data))).reset_index()
null_percent_categorical.columns = ['column_name', 'percent_of_null']
null_percent_categorical
```

```python
catagorical_data_null = round(null_percent_categorical[null_percent_categorical['percent_of_null']>0.85],2)
catagorical_data_null
```

```python
drop_columns_null = list(catagorical_data_null['column_name'])
print(drop_columns_null)
```

```python
drop_list.extend(drop_columns_null)
drop_list
```

Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest

Note that there are a few null values in the column. We will first replace those null values using the code below:


```python
categorical_data['GENDER'].value_counts()
```

```python
categorical_data['GENDER'].isna().sum()
```

```python
print(categorical_data['GENDER'].value_counts())
categorical_data['GENDER'] = categorical_data['GENDER'].fillna('F')
```

```python
categorical_data['GENDER'].value_counts()
```

```python
categorical_data['GENDER'] = categorical_data['GENDER'].replace(['J'],'other')
categorical_data['GENDER'] = categorical_data['GENDER'].replace(['A'],'other')
categorical_data['GENDER'] = categorical_data['GENDER'].replace(['C'],'other')
categorical_data['GENDER'] = categorical_data['GENDER'].replace(['U'],'other')
```
